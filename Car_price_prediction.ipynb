```python
def dropCol(df, cut = 0.9) :
        
    # Get correlation matrix and upper triangle
    corr_mtx = df.corr().abs()
    avg_corr = corr_mtx.mean(axis = 1)
    # up is a upper triangle 
    up = corr_mtx.where(np.triu(np.ones(corr_mtx.shape), k=1).astype(np.bool))
    
    # list of drop col 
    drop = list()

    for row in range(len(up)-1):
        col_idx = row + 1
        # row != col 
        for col in range (col_idx, len(up)):
            if(corr_mtx.iloc[row, col] > cut):
                # We choose attribute have its mean of corr larger
                if(avg_corr.iloc[row] > avg_corr.iloc[col]): 
                    drop.append(row)
                else: 
                    drop.append(col)
    
    # Make unique for element in drop 
    drop_set = list(set(drop))
    # dropcols_idx = drop_set
    dropcols_names = list(df.columns[[item for item in drop_set]])
    
    return(dropcols_names)
```

```python
dropCol(data, 0.8)
```

```python
import pickle 
# Store r2_score of the best model 
max_r2_score = 0
# Store mse of the best model
mse = 0
# Store the cutoff value of the best model 
cutoff = 0
# Drop list 
drop_l = None

# Log file 
log_file = open("train_log.txt", "w")

# Use RobustScaler: better for oulier 
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score, mean_squared_error


for i in np.arange(0.5, 1, 0.01):
    df1 = data1.copy()
    drop_list = dropCol(df1, i)
    if 'price' in drop_list:
        drop_list.remove('price')
    df1 = df1.drop(drop_list, axis = 1)
    
    X = df1.drop(columns = ['price']).copy()
    y = df1['price']

    transformer = RobustScaler().fit(X)
    X = transformer.transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(X,y , train_size = 0.8, random_state = 8)
    regr = linear_model.LinearRegression()

    # Train the model using the training sets
    regr.fit(X_train, y_train)
    
    # Make predictions using the testing set
    y_pred = regr.predict(X_test)
    
    """
    # The coefficients
    print("Coefficients: \n", regr.coef_)
    # The mean squared error
    print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
    # The coefficient of determination: 1 is perfect prediction
    print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
    """
    
    if max_r2_score < r2_score(y_test, y_pred):
        max_r2_score = r2_score(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        cutoff = i
        drop_l = drop_list
        filename = 'car_price_model.sav'
        pickle.dump(regr, open(filename, 'wb'))
        print("Saved")
    print(i, mean_squared_error(y_test, y_pred), r2_score(y_test, y_pred))
        
log_file.write("r2_score for best model: {}\n".format(max_r2_score))
log_file.write("mse for best model: {}\n".format(mse))
log_file.write("drop list for best model: {}\n".format(drop_l))
log_file.write("cutoff for best model: {}\n".format(cutoff))
log_file.close()
```
